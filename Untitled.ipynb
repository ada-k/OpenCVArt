```python
import random
import math

import numpy as np

import cv2
import scipy
import scipy.spatial
from sklearn.cluster import KMeans

from matplotlib import pyplot as plt
```

```python
# images path
path = "images/"
```

```python
def show_rgb_image(
    image,
    title=None,
    conversion=cv2.COLOR_BGR2RGB
):

    # Converts from one colour space to the other. this is needed as RGB
    # is not the default colour space for OpenCV
    image = cv2.cvtColor(image, conversion)

    # Show the image
    plt.figure(figsize = (8,7))
    plt.imshow(image)

    # remove the axis / ticks for a clean looking image
    plt.xticks([])
    plt.yticks([])

    plt.title(title)

    plt.show()
```

### Watercolor Effect

```python
img = cv2.imread(path + "img_2.jpg")
# sigma_s controls the size of the neighborhood. Range 1 - 200
# sigma_r controls the how dissimilar colors within the neighborhood will be averaged.
# A larger sigma_r results in large regions of constant color. Range 0 - 1
res_water = cv2.stylization(img, sigma_s=200, sigma_r=1)

# show_rgb_image(res_water, "3 in 1")

plt.figure(figsize=[20, 20])
plt.subplot(131)
plt.imshow(res_water[:, :, ::-1])
plt.title("Final Image")
plt.axis('off')
plt.subplot(132)
plt.imshow(img[:, :, ::-1])
plt.title("Original")
plt.axis('off')
```

### Pencil Sketch: Coloured and B&W

```python
img = cv2.imread(path + "img_16.jpg")
dst_gray, dst_color = cv2.pencilSketch(img,
                                       sigma_s=60,
                                       sigma_r=0.07,
                                       shade_factor=0.05)

# sigma_s and sigma_r are the same as in stylization.
# shade_factor is a simple scaling of the output image intensity.
# The higher the value, the brighter is the result. Range 0 - 0.1
plt.figure(figsize=[20, 20])
plt.subplot(131)
plt.imshow(dst_color)
plt.title("Final Image: Colored")
plt.axis('off')
plt.subplot(132)
plt.imshow(img[:, :, ::-1])
plt.title("Original")
plt.axis('off')
```

### Pointilist Art

```python
def compute_color_probabilities(pixels, palette):
    
    distances = scipy.spatial.distance.cdist(pixels, palette)
    maxima = np.amax(distances, axis=1)
    distances = maxima[:, None] - distances
    summ = np.sum(distances, 1)
    distances /= summ[:, None]
    
    return distances


def get_color_from_prob(probabilities, palette):
    
    probs = np.argsort(probabilities)
    i = probs[-1]
    
    return palette[i]


def randomized_grid(h, w, scale):
    
    assert (scale > 0)
    r = scale//2    
    grid = []
    
    for i in range(0, h, scale):
        
        for j in range(0, w, scale):
            
            y = random.randint(-r, r) + i
            x = random.randint(-r, r) + j    
            grid.append((y % h, x % w))    
            random.shuffle(grid)
            
    return grid


def get_color_palette(img, n=20):
    
    clt = KMeans(n_clusters=n)
    clt.fit(img.reshape(-1, 3))
    
    return clt.cluster_centers_


def complement(colors):
    
    return 255 - colors

def create_pointillism_art(img, primary_colors):
        
    radius_width = int(math.ceil(max(img.shape) / 1000))
    palette = get_color_palette(img, primary_colors)
    complements = complement(palette)
    palette = np.vstack((palette, complements))
    canvas = img.copy()    
    grid = randomized_grid(img.shape[0], img.shape[1], scale=3)
    pixel_colors = np.array([img[x[0], x[1]] for x in grid])
    color_probabilities = compute_color_probabilities(pixel_colors, palette)
    
    for i, (y, x) in enumerate(grid):
        color = get_color_from_prob(color_probabilities[i], palette)
        cv2.ellipse(canvas, (x, y), (radius_width, radius_width), 0, 0, 360, color, -1, cv2.LINE_AA)    
        
    return canvas


img = cv2.imread(path + "img_16.jpg")
res = create_pointillism_art(img,  20)
    
    
```

```python
plt.figure(figsize=[20, 20])
plt.subplot(131)
plt.imshow(res)
plt.title("Final Image")
plt.axis('off')
plt.subplot(132)
plt.imshow(img[:, :, ::-1])
plt.title("Original")
plt.axis('off')
```
